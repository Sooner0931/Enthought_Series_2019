## Problem 2: Multiple variables are stored in one column

```python
ebola = pd.read_csv('country_timeseries.csv')
```

```python
ebola.head(3)                   # you see that case/death count and country name was stored in a single column respectively
```

```python
ebola_melt = ebola.melt(id_vars = ['Date', 'Day'], var_name = 'case__death_country', value_name = 'num_of_counts')
```

```python
ebola_melt.iloc[[-1, 1]]
```

```python
ebola_melt['case__death_country']
```

```python
# my try:

for index, item in enumerate(ebola_melt['case__death_country']):
    case_or_death, country = item.split("_")
    ebola_melt['case_death'].loc[index] = case_or_death
    ebola_melt['country_name'].loc[index] = country
```

```python
ebola_melt.head(3)
```

```python
# now using pandas .str accessor
```

```python
ebola_melt.drop(['case_death', 'country_name'], axis = 1, inplace = True)
```

```python
ebola_melt.head(3)
```
